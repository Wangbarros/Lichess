# Getting and parsing data

```python
def mate_to_analysis(mate_in):
    ## When the computer analysis identifies a "Mate in X Moves", it just outputs the number of moves for a mate
    ## instead of an analysis in centipawns. This function returns a result of 1000 or -1000 according to which colour
    ## has a possible "mate in X moves"
    if mate_in <0:
        return -1000
    elif mate_in > 0:
        return 1000
    else:
        ## It doesnt make sense for a situation to be a "mate in 0 moves"
        raise ValueError('Game seems to be a mate in 0 moves. This makes no sense.')
        return 0
def truncate_centipawns(value):
    # Get value in centipawns and floors it to a max/min
    if value>1000:
        value = 1000
    elif value<-1000:
        value = -1000
    return value
```

```python
#Variables availible from game data json

ids = []
black_player = []
black_rating = []
white_player = []
white_rating = []
winner = []
status = []
created_at = []
last_at = []
game_turns = []


FMT = '%Y-%m-%d %H:%M:%S'

#The last 36 games are old games and won't be considered in the analysis
for game_number in range(1,total_number_games-36):
#for game_number in range(1,10):
    time.sleep(5)
    if game_number % 10 == 0:
        print(game_number)
    game_string_number = requests.get(game_string + str(game_number)).json()
    # Getting only rapid games, not correspondence (those with daysperturn)
    if 'daysPerTurn' not in game_string_number['currentPageResults'][0].keys():
        # Getting only games with 3 minutes per player
        # Getting only games with more than 3 moves (less than that are aborted games)
        # Getting only games in relevant period
        if (game_string_number['currentPageResults'][0]['clock']['initial'] == initial_time and 
            game_string_number['currentPageResults'][0]['clock']['increment'] == increment and
            game_string_number['currentPageResults'][0]['turns'] >= 3 and
            '2017' in time.strftime('%Y-%m-%d %H:%M:%S', 
                          time.gmtime(game_string_number['currentPageResults'][0]['lastMoveAt']/1000))):
            ids.append(game_string_number['currentPageResults'][0]['id'])
            black_player.append(game_string_number['currentPageResults'][0]['players']['black']['userId'])
            black_rating.append(game_string_number['currentPageResults'][0]['players']['black']['rating'])
            white_player.append(game_string_number['currentPageResults'][0]['players']['white']['userId'])
            white_rating.append(game_string_number['currentPageResults'][0]['players']['white']['rating'])
            status.append(game_string_number['currentPageResults'][0]['status'])
            # Game can be won by black, white or a draw. Draws may be agreed or by stalemate
            if ((game_string_number['currentPageResults'][0]['status']) != "draw" and
               game_string_number['currentPageResults'][0]['status'] != "stalemate"):
                winner.append(game_string_number['currentPageResults'][0]['winner'])
            else:
                winner.append("draw")
            game_turns.append(game_string_number['currentPageResults'][0]['turns'])
            created_at.append( time.strftime(FMT, 
                                       time.gmtime(game_string_number['currentPageResults'][0]['createdAt']/1000)))
            last_at.append( time.strftime(FMT, 
                                       time.gmtime(game_string_number['currentPageResults'][0]['lastMoveAt']/1000)))
```

```python
game_df = pd.DataFrame(
    {'id': ids,
     'status': status,
     'game_turns': game_turns,
     'created_at': created_at,
     'last_at': last_at,
     'winner': winner,
     'white_player': white_player,
     'black_player': black_player,
     'white_rating': white_rating,
     'black_rating': black_rating,
    })

def rating_dif(game_df):
    #Gets the rating difference between me and opponent. Negative if the oppnent has more rating.
    if game_df['white_player'] == "wangbarros":
        return game_df['white_rating'] - game_df['black_rating']
    else:
        return game_df['black_rating'] - game_df['white_rating']
def did_I_win(game_df):
    if game_df['winner'] == 'white':
        if game_df['white_player'] == 'wangbarros':
            return 'win'
        else:
            return 'loss'
    elif game_df['winner'] == 'draw':
        return 'draw'
    else:
        if game_df['black_player'] == 'wangbarros':
            return 'win'
        else:
            return 'loss'
def opponent_name(game_df):
        if game_df['white_player'] == "wangbarros":
            return game_df['black_player']
        if game_df['black_player'] == "wangbarros":
            return game_df['white_player']

def my_rating(game_df):
    if game_df['white_player'] == "wangbarros":
        return game_df['white_rating']
    if game_df['black_player'] == "wangbarros":
        return game_df['black_rating']
    
def opponent_rating(game_df):
    if game_df['white_player'] == "wangbarros":
        return game_df['black_rating']
    if game_df['black_player'] == "wangbarros":
        return game_df['white_rating']
    
def game_time(game_df):
    begin = datetime.datetime.strptime(game_df['created_at'], FMT)
    end = datetime.datetime.strptime(game_df['last_at'], FMT)
    game_df['game_time'] = (end - begin).seconds
    return game_df['game_time']

    
game_df['rating_dif'] = game_df.apply(lambda row: rating_dif(row), axis=1)
game_df['did_I_win'] = game_df.apply(lambda row: did_I_win(row), axis=1)
game_df['opponent_name'] = game_df.apply(lambda row: opponent_name(row), axis=1)
game_df['my_rating'] = game_df.apply(lambda row: my_rating(row), axis=1)
game_df['opponent_rating'] = game_df.apply(lambda row: opponent_rating(row), axis=1)
game_df['game_time'] = game_df.apply(lambda row: game_time(row), axis=1)

print((game_df['game_time'][1]))
```

```python
game_string = 'https://en.lichess.org/api/game/'
conditions_string = '?with_moves=1&with_analysis=1&with_movetimes=1&with_opening=1&with_moves=1'

#Variables availible from detailed game data json

times_black = []
times_white = []
analysis_white = []
analysis_black = []
white_moves = []
black_moves = []
opening = []

for index in range(0,len(ids)):
#for index in range(0,2):
    time.sleep(5)
    one_game = requests.get(game_string + ids[index] + conditions_string)
    times_black.append((one_game.json()['players']['black']['moveCentis']))
    times_white.append((one_game.json()['players']['white']['moveCentis']))
    opening.append((one_game.json()['opening']['name']))
    
    game_analysis_white = []
    game_analysis_black = []
    game_black_moves = []
    game_white_moves = []
    
    for turns in range(0,len((one_game.json()['moves']).split(" "))):
        #White moves on even turns
        if turns % 2 == 0:
            game_white_moves.append((one_game.json()['moves']).split(" ")[turns])
            
            #Check if the move was a mate. If it was a mate, there is no analysis
            if '#' not in (one_game.json()['moves']).split(" ")[turns]:
            
                if ('eval' in (one_game.json()['analysis'][turns].keys())):
                    game_analysis_white.append(truncate_centipawns(one_game.json()['analysis'][turns]['eval']))
                elif ('mate'in (one_game.json()['analysis'][turns].keys())):
                    game_analysis_white.append(mate_to_analysis(one_game.json()['analysis'][turns]['mate']))
                else:
                    print("No analysis and no mate?")
            else:
                #Checkmate by white/black will have the same number as "mate in X"
                game_analysis_white.append(1000)
        else:
            game_black_moves.append((one_game.json()['moves']).split(" ")[turns])
            
            #Check if the move was a mate. If it was a mate, there is no analysis
            if '#' not in (one_game.json()['moves']).split(" ")[turns]:
                if ('eval' in (one_game.json()['analysis'][turns].keys())):
                    game_analysis_black.append(truncate_centipawns(one_game.json()['analysis'][turns]['eval']))
                elif ('mate'in (one_game.json()['analysis'][turns].keys())): 
                    game_analysis_black.append(mate_to_analysis(one_game.json()['analysis'][turns]['mate']))
                else:
                    print("No analysis and no mate?")
            else:
                #Checkmate by white/black will have the same number as "mate in X"
                game_analysis_black.append(-1000)
                
    analysis_white.append(game_analysis_white)
    analysis_black.append(game_analysis_black)
    white_moves.append(game_white_moves)
    black_moves.append(game_black_moves)
    if index % 10 == 0:
        print(index)

one_game = requests.get(game_string + ids[0] + conditions_string)
```

```python
##Converting months days and hours from format "2" to "02" for sorting purposes
game_df['created_at'] = pd.to_datetime(game_df['created_at'])

def correct_day(game_df):
    if int(game_df['day'])<10:
        return '0' + str(game_df['day'])
    else:
        return game_df['day']
    
def correct_month(game_df):
    if int(game_df['month'])<10:
        return '0' + str(game_df['month'])
    else:
        return game_df['month']
    
def correct_hour(game_df):
    if int(game_df['hour'])<10:
        return '0' + str(game_df['hour'])
    else:
        return str(game_df['hour'])

game_df['day'] = (game_df["created_at"].dt.day)
game_df['month'] = (game_df["created_at"].dt.month)
game_df['hour'] = (game_df["created_at"].dt.hour)

game_df['day'] = game_df.apply(lambda row: correct_day(row), axis=1)
game_df['month'] = game_df.apply(lambda row: correct_month(row), axis=1)
game_df['hour'] = game_df.apply(lambda row: correct_hour(row), axis=1)

game_df['month-day'] = ((game_df['month'].map(str)) +'-' + (game_df['day'].map(str)))

#sorting month-day
order = list(set(reversed(game_df['month-day'])))
order = sorted(order)

fig = plt.figure(figsize=(10,10))

sns.plt.title('Result vs Month Day')

sns.countplot(x='month-day',data = game_df, hue = "did_I_win", order = order, 
             palette = {'win': 'green', 'draw': 'blue', 'loss': "red"}, hue_order=['win', 'draw','loss'],
             );

```

```python
def count_opening(opening, game_df):
    counter = 0
    for index in range(0,len(game_df['opening'])):
        if opening == game_df['opening'][index]:
            counter = counter+1
    return counter



def sub_opening(opening, game_df):
    
    for index in range(0,len(game_df['opening'])):
        
        if opening in game_df['opening'][index]:
            game_df['opening'][index] = opening
            
        elif ('Horwitz Defense' in game_df['opening'][index] or 
                "Zukertort Opening" in game_df['opening'][index] or
                "Nimzo-Larsen Attack" in game_df['opening'][index]):
            game_df['opening'][index] = 'Dutch Defense'
            
        elif "Three Knights" in game_df['opening'][index] or "Ruy Lopez" in game_df['opening'][index]:
            game_df['opening'][index] = "Four Knights Game"
            
        elif "Latvian" in game_df['opening'][index]:
            game_df['opening'][index] = "King's Pawn Game"
            
    return game_df

game_df['opening'] = opening

relevant_openings = ['Sicilian Defense','Dutch Defense','Caro-Kann','French Defense: Advance','French Defense: Winawer',
                    'French Defense: Tarrasch','French Defense: Exchange',
                    'Philidor Defense','Four Knights Game','Scandinavian Defense',"King's Pawn Game",
                    'French Defense: Other']

for x in relevant_openings:
    game_df = sub_opening(x, game_df)
    

francesas = ['French Defense: Advance','French Defense: Winawer','French Defense: Tarrasch','French Defense: Exchange']

for index in range(0,len(game_df['opening'])):
    
    if('French Defense: Pelikan' in game_df['opening'][index]):
        game_df['opening'][index] = 'French Defense: Advance'
        
    elif('French Defense: Knight' in game_df['opening'][index]):
        game_df['opening'][index] = 'French Defense: Exchange'
    
    elif ('French Defense' in game_df['opening'][index] and game_df['opening'][index] not in francesas):
        game_df['opening'][index] = 'French Defense: Other'
    elif game_df['opening'][index] not in relevant_openings:
        game_df['opening'][index] = "Other"

for x in relevant_openings:
    print(count_opening(x, game_df))

print(list(set(game_df['opening'])))
```
