```python
import yaml
import pprint
import requests
import time
import datetime
```

```python



initial_time = 180
increment = 0

one_game = requests.get('https://en.lichess.org/api/user/Wangbarros/games?nb=1&page=1')


one_game_dict = one_game.json()
one_game_status = one_game.status_code

total_number_games = one_game_dict['nbResults']
game_string = "https://en.lichess.org/api/user/Wangbarros/games?nb=1&page="

ids = []

for game_number in range(1,total_number_games-36):
#for game_number in range(1,16):
    print(game_number)
    game_string_number = requests.get(game_string + str(game_number)).json()
    if 'daysPerTurn' not in game_string_number['currentPageResults'][0].keys():
        if (game_string_number['currentPageResults'][0]['clock']['initial'] == initial_time and 
            game_string_number['currentPageResults'][0]['clock']['increment'] == increment and
            '2017' in time.strftime('%Y-%m-%d %H:%M:%S', 
                          time.gmtime(game_string_number['currentPageResults'][0]['lastMoveAt']/1000))):
            ids.append(game_string_number['currentPageResults'][0]['id'])



```
